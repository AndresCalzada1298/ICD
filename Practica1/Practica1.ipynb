```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

Primero se selecciona un DataSet de Calidad de los vinos 

```python
#importamos nuestra base de datos 
df = pd.read_csv('final_data.csv')  #cargamos el dataset llamado winequeality-red y lo asignamos a la variable df
print(df)
```

```python
df.head(10)   #Solamente nos quedamos con los primeros 10 elementos de todo el dataset 
```

A continuación se muestra la gráfica del histograma de cada atributo para toda la base de datos. En la cual es claro observar 
que para los atributos de total de aplicantes y numero de empleados se tiene una mayor cantidad de datos.

```python
#Aqui se gráfica el histograma de todo la base de datos, usamos el comando figsize(15,20) para que esto se vea más separado
#Usando color para cambiar el color de las gráficas a color rojo 
df.hist(figsize = (15,20), color = 'r');
```

Para la siguiente figura podemos observar como existe una relacion entre el número de empleados, donde se muestra que la densidad para  
los valores de 0 y 10000 es mayor.

```python
#Gráfica con jointplot 

sns.set_theme(style="ticks")
sns.jointplot(data = df, x = 'Total_applicants', y = 'Employee_count', kind="hex", color="#4CB391");
```

A continuación podemos observar la gráfica donde se muestra la dispersión de los datos para numero de empleados y el total 
de aplicantes para lo datos de PANDAS, es decir que para los numeros de empleados y el toal de aplicantes cuales son los que
saben o conocen el tema de uso de PANDAS

```python
#Hacemos una Gráfica con scatterplot

sns.scatterplot(data=df, x="Employee_count", y="Total_applicants", hue="PANDAS").set_title('Gráfica de tipo scatter plot para hacer el match entre el numero de empleados y el total de aplicantes');
```

Para el siguiente gráfico podemos observar que para el total de applicantes las frecuencias en el histograma se encuentran en valores bajos,
por lo que se puede mencionar que estos aplicantes no coinciden con los atributos para programación que se piden, por ejemplo
c++, java, python etc.

```python
#Histograma 
sns.histplot(data=df, x="Total_applicants", binwidth=3).set_title("Gráfica del histograma para el total de aplicantes");
```

Para el siguiente gráfico podemos observar que para número de empledos las frecuencias en el histograma coinciden en la mayoria para los atributos que se requieren para un empleo, por ejemplo, panda c++, java, python etc.

```python
#Grafica del histograma pero enfocado en el eje y
sns.histplot(data=df, y="Employee_count").set_title("Gráfica del histograma para los datos de número de empleados");
```

Para la siguiente gráfica se puede interpretar que para el total de aplicantes donde esta la mayor densidad de aplicantes que conocen o saben el manjedo de python.

```python
sns.violinplot(data=df, x="Total_applicants", y="PYTHON"); #Gráfica de tipo 
```

# Gráficas con MatplotLib

En la siguiente gráfica podemos observar que para más del 60% de los seguidores coinciden con las caracteríscas para poder aplicar para 
uno de los puestos 

```python
total = df['Total_applicants']
foll = df['LinkedIn_Followers']

plt.rcParams["figure.figsize"] = (8,12)       #Este comando se esa para aumentar el tamaño de la gráfica
fig, ax = plt.subplots()

ax.scatter(total, foll, c = 'b', edgecolors = 'green', marker=r'$\clubsuit$')  #Se hace la gráfica de los elemtos total y foll, asi como le elección de color 
                                                                               # y el tipo de figura con el cual se realiza la gráfica 
plt.title('Gráfica del total de aplicantes vs seguidores de linkedln')         #Se asigna el título de la gráfica 
plt.xlabel('Total de aplicantes')                                              #Nombre del eje x
plt.ylabel('Total de seguidores');                                             #Nombre del eje y
```

En la siguiente gráfica se puede notar que ára el total de applicantes cuales son los perfiles que coinciden para los empledos 
es decir se puede notar una correlación para los perfiles entre el número de empledos y los aplicantes 

```python
total = df['Total_applicants']
emplo = df['Employee_count']
fig, ax = plt.subplots()
ax.bar(total, emplo, width=1, edgecolor="white", linewidth=0.7)
plt.title('Gráfica del total de aplicantes vs cuenta de empleados')         #Se asigna el título de la gráfica 
plt.xlabel('Total de aplicantes')                                           #Nombre del eje x con el total de aplicantes
plt.ylabel('Número de empleados');                                          #Nombre del eje y con el numero de empleados 
```
