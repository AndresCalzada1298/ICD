```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.axes as ax
import seaborn as sns
from sklearn.feature_selection import mutual_info_classif
from sklearn.ensemble import RandomForestClassifier
```

```python
#Cargamos nuestro DataSet
data = pd.read_csv('full.csv')
data.head()
```

```python
#Procedemos a verificar si existen valores nulos NaN dentro de nuestro DataFrame
#Al observaqr la información de nuestro DataFrame podemos observar que existen clases que tienen estos valores
data.info()
```

Por medio de la función sns.heatmap() vamos a realizar una representación gráfica de los valores nulos que aparacen en nuestro DataFrame, estos valores que NaN son mostrados en el gráfico con el colo más claro que corresponde al valor 1, es decir el valor booleano que nos comfirma la existenacia de estos valores en el DataFrame.
Así es más claro visualizar dentro del gráfico que los valores NaN se encuentran tanto espaciados como de manera continua en diferentes clases del DaraFrame.

```python
sns.heatmap(data.isna());
```

```python
#Vamos a realizar una suposición, en la cual vamos a mencionar que uno de las clases importantes dentro de nuetro DataFrame es la columna de edad
#Por lo que vamos a realizar el histograma de la edad con respecto al DataFrame  
sns.histplot(data=data, x="Age", binwidth=3).set_title("Gráfica del histograma para las edades");
```

```python
#Vamos a proceder a eliminar los valores nulos con la siguiente función
data = data.fillna(0)
#Obtenemos nuevamente la información y podemos observar que efectivamente los valores nulos han sido eliminados
data.info()
```

```python
#Mostramos las columnas para conocer donde esta nuestra clase de interés, en este caso vamos a seleccionar la columna 'Survived' como nuestra clase
data.columns
```

```python
#Vamos a proceder a colocar nuestra clase al final de nuestro DataFrame 
column = data.pop('Survived')
data.insert(20, 'Survived', column)
data.info()
```

```python
#Una vez que ya recorrimos nuestra clase o nuestra columna de importacia procedemos a realizar una gráfica inicial de como es el comportamiento 
data.hist(figsize = (10,10), color = '#290DF7');
```

```python
#Histograma 
sns.histplot(data=data, x="Age", binwidth=3).set_title("Gráfica del histograma para las edades");
```

#### **Observación**

Es fácil notar que en la gráfica de edades la mayoria de datos se inclinaron hacia el valor de cero, es decir pasajeros con edades de cero años
pero es importante notar que al momento de hacer que los valores de NaN fueran reemplazados los datos faltantes se reemplazaron por ceros, por lo que esta gráfica nos quiere decir que existian muchos valores nulos 

#### **Observación 2**

Otra observación con respecto a este Dataframe es que las columnas que contengan nombres en este caso para nosotros no es algo relevante
ya que no es un factor para determinar si las personas sobrevivieron o no, por lo que vamos a proceder a eliminarla 


```python
#Eliminamos los las columnas que no nos generan importancia para conocer si las personas sobrevivieron al titanic 
#Como parte de la limpieza de datos 
data = data.drop(['Name', 'Ticket', 'Name_wiki','Hometown', 'Boarded', 'Destination', 'Sex','Cabin','Embarked','Lifeboat','Body'], axis = 1)
```

```python
#Hacemos la separación de atributos y a la clase una vez que se tiene a la clase hasta el final del DataFrame
X = data.iloc[:, 0:-1]   #Separamos a los atributos 
y = data.iloc[:, -1]     #Separamos a la clase
```

```python
#Observamos la ganancia de información de cada atributo con respecto a la clase 
X_nueva = mutual_info_classif(X, y)
feat_imp = pd.Series(X_nueva, data.columns[0:len(data.columns)-1])
feat_imp.plot(title="Características con mayor importancia", kind='barh')
plt.show()
```

```python
#Ahora usamos la función de RandomForestClassifier para conocer cuales son los atributos que tienen una mayor importancia 

model = RandomForestClassifier(n_estimators=340)
model.fit(X,y)
importances = model.feature_importances_

feat_df = pd.DataFrame({"Features": pd.DataFrame(X).columns, "Importancia": importances})
feat_df.set_index('Importancia')
feat_df.plot.barh(title="Características con mayor importancia" ,color = '#290DF7');
```

Donde es facil observar que el atributo que tiene mayor importancia en nuestro DataFrame es:
1. Passerngerid
2. Fare
3. Wikiid
4. Parch 


#### **Reducción de dimensionalidad**

```python
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
```

```python
X_PCA = PCA(n_components=2).fit_transform(X)

scatter = plt.scatter(X_PCA[:,0], X_PCA[:,1], c=y)
plt.title('PCA plot in 2D') 
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.legend(*scatter.legend_elements())
plt.show()
```

```python
X_TSNE = TSNE().fit_transform(X) 

scatter = plt.scatter(X_TSNE[:,0], X_TSNE[:,1], c=y)
plt.title('t-SNE plot in 2D')
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.legend(*scatter.legend_elements())
plt.show()
```

#### **Regresión lineal Extra**

```python
data_new = data[['Fare', 'Age']]
data_new.head()

data_new.shape
```

```python
#Parte de los datos  los primeros son para entrenamiento 
train_input = np.array(data_new.Age[0:870]).reshape(870,1)
#Los segundos son para la prediccion 
train_output  = np.array(data_new.Fare[0:870]).reshape(870,1)
```

```python
# valid dataset and labels
test_input = np.array(data_new.Age[870:1310]).reshape(439,1)
test_output  = np.array(data_new.Fare[870:1310]).reshape(439,1)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(train_input, train_output)
print(reg.score(train_input, train_output))
```

```python
#Prediction on test data
print(reg.coef_)
print(reg.intercept_)
y_pred = reg.predict(test_input)
```

```python
# Plot the regression line with actual data pointa
plt.plot(test_input, test_output, '+', label='Actual values')
plt.plot(test_input, y_pred, label='Predicted values')
plt.xlabel('Test input')
plt.ylabel('Test Output or Predicted output')
plt.legend()
plt.show()
```
